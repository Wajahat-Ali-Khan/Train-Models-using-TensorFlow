```python
b = pd.read_csv("sonar.csv")
# b.data
```

```python
b.shape
b.head()
```

```python
#label encoding

enc = pp.LabelEncoder()
b.iloc[:,60] = enc.fit_transform(b.iloc[:, 60])
b.head()
```

```python
#for shuffle data
b.sample(frac=1)
```

```python
#split data into train and test
x_train, x_test, y_train, y_test = train_test_split(b.iloc[:, :60], b.iloc[:, 60], test_size=0.2)
```

```python
#split data into train and validation
# train, val = train_test_split(b, test_size=0.1, random_state=1)
x_train, x_val, y_train, y_val = train_test_split(b.iloc[:, :60],  b.iloc[:, 60], test_size=0.1,random_state=1)
```

```python
print('Train', x_train.shape)
print('Test', x_test.shape)
print('val',x_val.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
print('y_val shape:', y_val.shape)
```

```python
# one hot coding
one_hot_train=to_categorical(y_train)
# one_hot_train
one_hot_test=to_categorical(y_test)
# one_hot_test
one_hot_val=to_categorical(y_val)
# one_hot_val
```

```python
#train model
model = models.Sequential()
model.add(layers.Dense(32, activation='relu', input_shape=(60,)))
model.add(layers.Dense(16, activation='relu'))
# tf.keras.layers.Dropout(0.2),
model.add(layers.Dense(1, activation='sigmoid'))
from tensorflow.keras import optimizers
model.compile(optimizer=optimizers.RMSprop(lr=0.001),
loss='binary_crossentropy',
metrics=['accuracy'])
```
